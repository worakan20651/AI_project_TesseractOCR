```python
import os
import random
import pathlib
import subprocess

training_text_file = '../langdata/tha.training_text'

lines = []

with open(training_text_file, 'r', encoding="utf-8") as input_file:
    for line in input_file.readlines():
        lines.append(line.strip())

output_directory = '../tesstrain/data/PonyBold-ground-truth'

if not os.path.exists(output_directory):
    print(output_directory)
    os.mkdir(output_directory)

random.shuffle(lines)

line_count = 0
```

```python
line_count = 0
font_path = "/home/seenam/.fonts"
for line in lines:
    training_text_file_name = pathlib.Path(training_text_file).stem
    line_training_text = os.path.join(output_directory, f'{training_text_file_name}_{line_count}.gt.txt')
    with open(line_training_text, 'w', encoding="utf-8") as output_file:
        output_file.writelines([line])

    file_base_name = f'tha_{line_count}'

    subprocess.run([
        'text2image',
        f'--fonts_dir={font_path}',
        '--font=PonyBold',
        f'--text={line_training_text}',
        f'--outputbase={output_directory}/{file_base_name}',
        '--xsize=3600',
        '--ysize=480',
        '--unicharset_file=langdata/tha.unicharset'
    ])
    # result = subprocess.run(text2image_command)
    # if(result.returncode == 0):
    #         print(f'Created .box and .tif files for {file_base_name}')

    line_count += 1

print('Processing completed.')
```
