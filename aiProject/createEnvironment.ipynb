```python
print("start process")
# Specify the path to the Tesseract executable (change as needed)
tesseract_path = 'tesseract'
lang = '-l tha'

# Directory containing the images
image_folder = '../y19_r32'

# Output directory for .box files
output_dir = '../testdir'

#specify language with thai
language = 'tha'
dpi = 300
```

```python
def numeric_part(filename):
    # Extract the numerical part (e.g., 0001) from the filename
    numeric_str = filename.split('-')[1].split('.')[0]
    # Convert to an integer for sorting
    return int(numeric_str)
```

```python
# Ensure the output directory exists
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# List all image files in the folder
image_files = [f for f in os.listdir(image_folder) if f.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp', '.tiff'))]

# Sort the list of filenames using the custom key function
image_files = sorted(image_files, key=numeric_part)
# print(image_files)

# Process each image to generate .box file
for image_file in image_files:
    image_path = os.path.join(image_folder, image_file)
    # print(image_path)
    output_box_file = os.path.join(output_dir, os.path.splitext(image_file)[0])

    if not os.path.exists(output_box_file):
        # Run Tesseract to create .box file
        tesseract_cmd = [
            'tesseract',    # Path to the Tesseract executable (adjust if necessary)
            image_path,    # Input image file
            output_box_file,     # Output box file
            '-l', language, # Specify the language as Thai
            '--dpi', str(dpi), #specify dpi of image
            '--psm', str(7),
            'batch.nochop', # Output in box format
            'makebox',      # Generate box files
        ]
        # print(tesseract_cmd)
        result = subprocess.run(tesseract_cmd)
        # print(subprocess.run(cmd))
        if(result == 0):
            print(f'Created .box file for {image_file}')

print('Processing completed.')
```
